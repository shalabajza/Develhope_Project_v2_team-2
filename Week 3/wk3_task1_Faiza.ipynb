```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
df = pd.read_csv('/content/1_total_installs_per_category.csv')
```

```python
#Qn2.plot a bar chart for categories with the total number of installing numbers in each category 
plot.figure()
f, ax = plot.subplots(figsize=(14,7))
#sns.set(font_scale=1.8)
g = sns.barplot(ax =ax, x='Category', y='total_installs', data=df)
plot.legend(['installs'])
plot.title("TOTAL PRICE CHART",weight = 'bold', fontsize= 'x-large')
plot.xlabel('CATEGORY', weight = 'bold')
plot.ylabel('TOTAL PRICE PER APP', weight = 'bold')
g.set_xticklabels(labels = df['Category'], rotation=80, fontsize = 9, weight = 'bold')
plot.show(g)
```

```python
#Qn3.plot a bar chart for the total prices of each paid app in each category ( the summation of all prices in the same category)
df1 = pd.read_csv('/content/2_total_app_price_per_category.csv')
```

```python
f, ax = plot.subplots(figsize=(14,7))
g = sns.stripplot(ax =ax, x='Category', y='total_price_per_app', data=df1, s = 15, c='brown')
sns.set_style('whitegrid')
plot.legend(['price'])
g.set_xticklabels(labels = df['Category'], rotation=80, fontsize = 9,weight = 'bold')
plot.title("TOTAL PRICE CHART",weight = 'bold', fontsize= 'x-large')
plot.xlabel('CATEGORY', weight = 'bold')
plot.ylabel('TOTAL PRICE PER APP', weight = 'bold')
plot.show(g)
```

```python
#Qn4.plot a bar chart of the total profit of each category by multiplying the price by the number of installs
df2 = pd.read_csv('/content/3_total_sales_per_category.csv')
```

```python
f, ax = plot.subplots(figsize=(14,7))
#sns.set(font_scale=1.8)
g = sns.pointplot(x='Category', y='total_profit', data=df2)
sns.set_style('darkgrid')
plot.legend(['profit'])
g.set_xticklabels(labels = df['Category'], rotation=80, fontsize = 9, weight = 'bold')
plot.title("TOTAL PROFIT CHART",weight = 'bold', fontsize= 'x-large')
plot.xlabel('CATEGORY', weight = 'bold')
plot.ylabel('TOTAL PROFIT', weight = 'bold')
plot.show(g)
```
