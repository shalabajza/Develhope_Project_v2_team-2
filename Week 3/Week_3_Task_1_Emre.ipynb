```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create dataframe out of csv file through pandas method
df = pd.read_csv("../Week 1/exported_data/googleplaystore_clean.csv")
```

```python
# Create new dataframe to plug in bar plot, grouping by category column and aggregating Installs
df_installplot = df.groupby("Category")["Installs"].sum().sort_values(ascending=False).reset_index()
df_priceplot = df.groupby("Category")["Price"].sum().sort_values(ascending=False).reset_index()
df["Revenue"] = df["Installs"] * df["Price"]
df_revenueplot = df.groupby("Category")["Revenue"].sum().sort_values(ascending=False).reset_index()
```

```python
sns.set_style("ticks")
plt.figure(figsize=(10,6))
sns.barplot(data=df_installplot, x="Category", y="Installs")
sns.set_context("talk", font_scale=0.7)
plt.xticks(rotation=90)
plt.title("App Install Totals per Category")
plt.ylabel("Installs (in Billions)")
plt.xlabel("App Categories")
plt.show()
plt.clf()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,6))
ax = sns.barplot(data=df_priceplot, y="Category", x="Price", orient="h")
label = ax.get_yticklabels()
label[0].set_color("green")
label[1].set_color("green")
label[2].set_color("green")
sns.set_context("poster", font_scale=0.4)
plt.title("App Price Totals per Category")
plt.xlabel("Price Totals (in USD)")
plt.ylabel("App Categories")
plt.show()
plt.clf()
```

```python
sns.set_style("whitegrid")
sns.catplot(data=df_revenueplot, x="Category", y="Revenue", kind="bar")
sns.set_context("paper", font_scale=1)
plt.xticks(rotation=90)
plt.title("App Revenue Totals per Category")
plt.ylabel("Revenue (in Billions)")
plt.xlabel("App Categories")
plt.show()
plt.clf()
```
