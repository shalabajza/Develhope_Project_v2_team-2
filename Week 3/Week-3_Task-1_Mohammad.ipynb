### Preprocessing

importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

importing the data

```python
df = pd.read_csv("../Week 1/exported_data/googleplaystore_clean.csv")
```

```python
df.tail()
```

### Obtaining the number of installs per category

Grouping the data according to the categories and summing the number of installs

```python
category_installs = df.groupby('Category')['Installs'].sum().sort_values(ascending=False)
category_installs.to_csv("plots_muhammed/installs.csv")
category_installs.head()
```

My most favorite plot:

```python
installs = pd.read_csv("plots_muhammed/installs.csv")
sns.set_theme(style="darkgrid")
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(x="Installs", y="Category", data=installs, palette="bright")
plt.xlabel("Number of Installs in Billions")
plt.ylabel("The Categories")
plt.title("Number of Installs Per Category")
```

2nd most favorite:

```python
sns.set_theme(style="darkgrid")
category_installs.plot(kind='bar', color='teal')
plt.xlabel("The Categories")
plt.ylabel("Number of Installs")
plt.title("Number of Installs Per Category")
plt.show()
```

### Obtaining the total profit per category

Importing the dataset to a new variable

```python
df_profit = pd.read_csv('../googleplaystore_clean.csv')
```

Creating a new column that contains the amount of profit per app

```python
df_profit['Profit'] = round((df_profit.Installs*df_profit.Price), 2)
```

grouping the data according to the category and summing the profit

```python
category_profit = df_profit.groupby('Category')["Profit"].sum().sort_values(ascending=False)
category_profit.to_csv("plots_muhammed/profit.csv")
category_profit.head()
```

My most favorite plot:

```python
profit = pd.read_csv("plots_muhammed/profit.csv")
sns.set_theme(style="darkgrid")
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(x="Profit", y="Category", data=profit, palette="bright")
plt.xlabel("Total Profit")
plt.ylabel("The Categories")
plt.title("Total Profit Per Category")
```

2nd most favorite:

```python
sns.set_theme(style="darkgrid")
category_profit.plot(kind='bar', color='teal')
plt.xlabel("The Categories")
plt.ylabel("Total Profit")
plt.title("Total Profit Per Category")
plt.show()
```
