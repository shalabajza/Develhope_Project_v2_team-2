importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

importing the data

```python
df = pd.read_csv('../googleplaystore_clean.csv')
```

```python
df.tail()
```

### Obtaining the top 5 installed apps in each category

```python
top_apps = df.groupby(['Category']).apply(lambda x: x.sort_values(["Installs"], ascending = False)).reset_index(drop=True)
# select top N rows within each continent
top_apps = top_apps.groupby('Category').head(5).reset_index(drop=True)
```

```python
top_apps.head(50)
```

```python
top_apps2 = df.groupby(['Category']).apply(lambda x: x.sort_values(["Installs", "Rating", "Reviews"], ascending = False)).reset_index(drop=True)
# select top N rows within each continent
top_apps2 = top_apps2.groupby('Category').head(5).reset_index(drop=True)
```

```python
top_apps2.head(50)
```

Plotting the results

```python
#sns.set_theme(style="darkgrid")
#sns.set(rc = {'figure.figsize':(10,6)})
top_apps2.plot(x='App', y='Installs', kind='bar', color='teal')
plt.xlabel("The Apps")
plt.ylabel("Number of Installs")
plt.title("Top 5 Most Installed Apps (For Each Ctegory)")
#plt.xticks(rotation=45)
plt.show()
```
