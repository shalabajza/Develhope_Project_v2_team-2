# TASK 2:
**print a table and a plot chart for top 5 download apps with smallest size and highest downloading number**

```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
df= pd.read_csv('/content/googleplaystore_clean.csv')
```

```python
df.head(5)
```

top 5 download apps with smallest size and highest downloading number

```python

top5_smallest_app= df.sort_values(['Size','Installs'],ascending=[True,False]).iloc[:5,0:6]
#top5_smallest_app=df.sort_values(['Size'],ascending=True).iloc[:5,0:6]
top5_smallest_app
```

We see that there is no much relationship between size and the number of installs. 

its plot:

```python
plot.figure()
f, ax = plot.subplots(figsize=(14,7))
g=sns.scatterplot(x= 'App', y='Installs', data=top5_smallest_app, s=100)
plot.legend(['downloads'])
plot.title("TOP 5 SMALLEST APP WITH THIER NO. OF. INSTALLS",weight = 'bold', fontsize= 'x-large')
plot.xlabel('Apps', weight = 'bold')
plot.ylabel('No.of.Installs(Millions)', weight = 'bold')
plot.xticks(rotation=0, fontsize = 9, weight = 'bold')
plot.show(g)
```

# TASK3
**Print a table and a plot chart for top 5 download apps in each category free and paid (if there is paid)**

```python
top5_app = df.sort_values(['Installs'],ascending=False).groupby(['Category','Type']).head(5)
top5_app 
```

```python
plot.figure()
f, ax = plot.subplots(figsize=(20,10))
sns.set_style('darkgrid')
g=sns.scatterplot(x= 'Category', y='Installs', data=top5_app, size='Installs', sizes=(15,80))
plot.legend(['downloads'])
plot.title("TOP 5 APP IN EACH CATEGORY",weight = 'bold', fontsize= 'x-large')
plot.xlabel('Category', weight ='bold')
plot.ylabel('No.of.Installs(Billions)', weight = 'bold')
plot.xticks(rotation=90, fontsize = 9, weight = 'bold')
plot.show(g)
```
