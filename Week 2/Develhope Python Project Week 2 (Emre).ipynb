```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create dataframe out of csv file through pandas method
df = pd.read_csv("googleplaystore_clean.csv")
```

```python
# Create new dataframe of unique category values by dropping duplicates in Category column, assign the same column as its own DF
df_category = df.drop_duplicates("Category", ignore_index=True)
df_category = df_category[["Category"]]
```

```python
# Create new dataframe to plug in bar plot, grouping by category column and aggregating Installs
df_installbarplot = df.groupby("Category")["Installs"].agg("sum")
```

```python
# Using pandas plotting, matplotlib labeling. Sorting values for more readable chart.
df_installbarplot.sort_values(ascending=False).plot(kind="bar")
plt.ylabel("Number of Installs (in Billions)")
plt.title("Number of Installs Per Category")
plt.show()
```

```python
# Same stuff as before, this time for price info.
df_pricebarplot = df.groupby("Category")["Price"].agg("sum")
df_pricebarplot.sort_values(ascending=False).plot(kind="bar")
plt.ylabel("Total Price ($)")
plt.title("Total Price Per Category")
plt.show()
```

```python
# Create new column in the original dataframe by multiplying 2 columns. Following same steps as two previous chart to generate the bar chart.
df["Revenue"] = df["Installs"] * df["Price"]
df_revenuebarplot = df.groupby("Category")["Revenue"].agg("sum")
df_revenuebarplot.sort_values(ascending=False).plot(kind="bar")
plt.ylabel("Total Revenue (In 100 Million Dollars)")
plt.title("Total Revenue Per Category")
plt.show()
```
