# Using Python Code

### Obtaining the number of installs per category

Grouping the data according to the categories and summing the number of installs

```python
category_installs = df.groupby(['Category']).agg({'Installs':['sum']})
category_installs
```

exporting the data frame to manually change the column names 

```python
category_installs.to_csv('category_installs.csv')
```

importing the edited data frame

```python
installs_sorted = pd.read_csv('category_installs.csv')
installs_sorted
```

sorting the number of installs from high to low

```python
installs_sorted = installs_sorted.sort_values(["Number of Installs"], ascending=False)
installs_sorted
```

```python
installs_sorted.plot(kind='bar', x="Category", y="Number of Installs", color='teal')
plt.xlabel("The Categories")
plt.title("Number of Installs Per Category")
plt.show()
```

### Obtaining the total app prices per category

grouping the data according to the categories and summing the price per app

```python
category_prices = df.groupby(['Category']).agg({'Price':['sum']})
category_prices
```

exporting the data frame to manually edit the column names

```python
category_prices.to_csv('category_prices.csv')
```

importing the edited data frame

```python
prices_sorted = pd.read_csv('category_prices.csv')
```

sorting the total app prices from high to low

```python
prices_sorted = prices_sorted.sort_values(["Total App Price"], ascending=False)
prices_sorted
```

Plotting the total price per category

```python
prices_sorted.plot(kind='bar', x="Category", y="Total App Price", color='teal')
plt.xlabel("The Categories")
plt.title("Total Price Per Category")
plt.show()
```

### Obtaining the total profit per category

Importing the dataset to a new variable

```python
df_profit = pd.read_csv('googleplaystore_clean.csv')
```

Creating a new column that contains the amount of profit per app

```python
df_profit['Profit'] = round((df_profit.Installs*df_profit.Price), 2)
```

grouping the data according to the category and summing the profit

```python
category_profit = df_profit.groupby(['Category']).agg({'Profit':['sum']})
category_profit
```

Exporting the data frame to manually edit the column names 

```python
category_profit.to_csv('category_profit.csv')
```

importing the edited data frame

```python
profit_sorted = pd.read_csv('category_profit.csv')
```

sorting the total profit from high to low

```python
profit_sorted = profit_sorted.sort_values(["Total Profit"], ascending=False)
profit_sorted
```

```python
profit_sorted.plot(kind='bar', x="Category", y="Total Profit", color='teal')
plt.xlabel("The Categories")
plt.title("Total Profit Per Category")
plt.show()
```
