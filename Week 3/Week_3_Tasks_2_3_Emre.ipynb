```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../Week 1/exported_data/googleplaystore_clean.csv")
df
```

```python
df_task1 = df.sort_values(["Size", "Installs"], ascending=[True, False]).head(5)
fig, ax = plt.subplots(figsize=(15, 6))
sns.scatterplot(ax = ax, data=df_task1, x="App", y="Size", size="Installs",sizes=(20, 400))
sns.set_context("talk", font_scale=0.7)
plt.xticks(rotation=60)
plt.title("5 Smallest Apps by Size with Highest Install Totals")
plt.ylabel("Size (in Mb)")
plt.xlabel("App Names")
plt.show()
```

```python
df_task2 = df.sort_values("Installs", ascending=False).groupby(["Category", "Type"]).head(5)
```

```python
sns.set_style("darkgrid")
sns.scatterplot(data=df_task2, x="Category", y="Installs", hue="Type", size="Installs", sizes=(20, 100), alpha=0.5)
sns.set(rc = {'figure.figsize':(20,6)})
plt.xticks(rotation=60)
plt.yscale("log")
plt.title("Top 5 Downloaded Apps per Category and Type")
plt.ylabel("Installs (in Billions, log scale)")
plt.xlabel("App Categories")
plt.show()
plt.clf()
```

```python
sns.set_style("darkgrid")
sns.stripplot(data=df_task2, x="Category", y="Installs", hue="Type", alpha=0.5)
sns.set(rc = {'figure.figsize':(15,8)})
plt.xticks(rotation=90)
plt.yscale("log")
plt.title("Top 5 Downloaded Apps per Category and Type")
plt.ylabel("Installs (in Billions, log scale)")
plt.xlabel("App Categories")
plt.show()
plt.clf()
```
