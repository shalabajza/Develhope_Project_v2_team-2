### Preprocessing

importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

importing the data

```python
df = pd.read_csv("../Week 1/exported_data/googleplaystore_clean.csv")
```

```python
df.tail()
```

Checking the relation between the content rating and the price 

```python
df.groupby(['Content_Rating'])['Price'].sum().sort_values(ascending=False).plot.bar(figsize=(10,6), color="teal")
plt.title("Total App Price Per Content Rating")
plt.ylabel("Total App Price")
plt.xlabel("Content Rating")
plt.show()
```

### More detailed view

Since we are dealing with The Price, I'll be dropping the free apps for now.

```python
df_paid = df[df["Type"] != "Free"]
df_paid.head()
```

### More detailed view

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_paid, x="Content_Rating", y="Price", hue="Content_Rating")
plt.title("Distribution of App Prices Per Content Rating")
plt.xlabel("Content Rating")
plt.legend(bbox_to_anchor=(1, 1))
```
