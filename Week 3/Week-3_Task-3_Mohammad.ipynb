importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

importing the data

```python
df = pd.read_csv('../googleplaystore_clean.csv')
```

```python
df.tail()
```

### Obtaining the top 5 installed apps in each category

```python
top_apps = df.groupby(['Category']).apply(lambda x: x.sort_values(["Installs"], ascending = False)).reset_index(drop=True)
top_apps = top_apps.groupby('Category').head(5).reset_index(drop=True)
```

```python
top_apps.head(10)
```

If you look at the table, you can see that even though Textgram and Canva have the same number of installs Textgram is listed above Canva. But here's the thing, if you check their ratings Canva has a heigher rating than Textgram. So that's why moving forward I'll be taking an extra step where in addition to the number of installs I'll also sort according to the rating and number of reviews. I know that we were not told to do so, but I fell it's a necessary step in order to obtain much better results.

```python
top_apps2 = df.groupby(['Category']).apply(lambda x: x.sort_values(["Installs", "Rating", "Reviews"], ascending = False)).reset_index(drop=True)
top_apps2 = top_apps2.groupby('Category').head(5).reset_index(drop=True)
```

```python
top_apps2.head(10)
```

### Splitting the dataset into each category and potting the results
