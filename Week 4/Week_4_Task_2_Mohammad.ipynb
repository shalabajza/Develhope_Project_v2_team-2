### Preprocessing

importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from afinn import Afinn
```

importing the data

```python
df = pd.read_csv("../Week 0 (Original database)/googleplaystore_user_reviews.csv")
df
```

Taking only the app names and the review from the original dataset

```python
df = df.loc[ : , ["App", "Translated_Review"]]
df.head()
```

```python
df.info()
```

Replacing missing values with empty strings

```python
df.fillna(' ', inplace=True)
df
```

```python
df.to_csv("googleplaystore_user_reviews_cleaned.csv")
```

### affin

Creating a list that contains only the reviews.

```python
reviews = df.loc[:,"Translated_Review"].str.lower()
reviews
```

Obtaining the scores

```python
afn = Afinn()
scores = [afn.score(review) for review in reviews]
```

adding the scores to the dataset

```python
df['affin_score'] = scores
df
```

### np

importing the negative word into a list

```python
n = pd.read_excel("data/n.xlsx", header=None, names=['words'])
n = n.words.to_list()
#n
```

importing the positive words into a list

```python
p = pd.read_excel("data/p.xlsx", header=None, names=['words'])
p = p.words.to_list()
#p
```

removing punctuation marks from the reviews for The NP evaluation.

```python
import string

reviews_np = df.loc[:,"Translated_Review"].str.lower().str.translate(str.maketrans('', '', string.punctuation))
reviews_np
```

obtaining the score by checking if a word is present in a list or not and scoring it accordingly

```python
np_scores = []
score = 0
for i in reviews_np:
    word_list = i.split()
    
    for j in word_list:
        if j in p:
            score = score+1
        if j in n:
           score = score-1
        else:
            continue
    np_scores.append(score)
    score = 0
    
#print(np_scores)
```

adding the np scores to the dataset

```python
df['np_score'] = np_scores
df
```
