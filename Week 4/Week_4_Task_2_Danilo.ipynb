```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from afinn import Afinn
```

```python
df = pd.read_csv("../Week 1/exported_data/googleplaystore_user_reviews_clean.csv")
```

```python
df
```

```python
df["Translated_Review"] = df["Translated_Review"].str.lower()
df = df[["App","Translated_Review"]]
```

```python
df
```

```python
afn = Afinn()
df['affin_scores'] = [afn.score(review) for review in df['Translated_Review']]
```

```python
n = pd.read_excel("Data/n.xlsx", header=None, names=['words'])
n = n.words.to_list()

p = pd.read_excel("Data/p.xlsx", header=None, names=['words'])
p = p.words.to_list()
```

```python
np_scores = []
for i in df['Translated_Review']:
    counter = 0
    word_list = i.split()
    for j in word_list:
        if j in p:
            counter = counter+1
        if j in n:
           counter = counter-1
    np_scores.append(counter)

df['np_scores'] = np_scores
```

```python
df = df.groupby("App")[["np_scores",'affin_scores']].sum().reset_index()
```

```python
df2 = pd.read_csv("../Week 1/exported_data/googleplaystore_clean.csv")
df_paid_only = df2[df2['Type'] == 'Paid']
df_paid_only
```
