### Preprocessing

importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from afinn import Afinn
```

importing the data

```python
df = pd.read_csv("../Week 0 (Original database)/googleplaystore_user_reviews.csv")
df
```

Taking only the app names and the review from the original dataset

```python
df = df.loc[ : , ["App", "Translated_Review"]]
df.head()
```

```python
df.info()
```

Replacing missing values with empty strings

```python
df.fillna('', inplace=True)
df
```

Creating a list that contains only the reviews

```python
reviews = df.loc[:,"Translated_Review"]
```

```python
afn = Afinn()
scores = [afn.score(review) for review in reviews]
```

```python
df['affin_scores'] = scores
df
```
