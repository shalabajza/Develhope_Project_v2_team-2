```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
#Qn2.plot a bar chart for categories with the total number of installing numbers in each category 
df = pd.read_csv('/content/category_installs.csv')
plot.figure()
f, ax = plot.subplots(figsize=(14,7))
g = sns.barplot (x= 'Category', y='Number of Installs', data=df)
plot.legend(['installs'])
plot.title("TOTAL INSTALL PER CATEGORY",weight = 'bold', fontsize= 'x-large')
plot.xlabel('CATEGORY', weight = 'bold')
plot.ylabel('TOTAL INSTALL PER APP', weight = 'bold')
g.set_xticklabels(labels = df['Category'], rotation=80, fontsize = 9, weight = 'bold')
plot.show(g)
```

```python
#Qn3.plot a bar chart for the total prices of each paid app in each category ( the summation of all prices in the same category)
df1 = pd.read_csv('/content/category_prices.csv')
f, ax = plot.subplots(figsize=(14,7))
g = sns.stripplot(x='Category', y='Total App Price', data=df1, s = 10, c='brown')
sns.set_style('whitegrid')
plot.legend(['price'])
g.set_xticklabels(labels = df['Category'], rotation=80, fontsize = 9,weight = 'bold')
plot.title("TOTAL PRICE CHART",weight = 'bold', fontsize= 'x-large')
plot.xlabel('CATEGORY', weight = 'bold')
plot.ylabel('Total App Price', weight = 'bold')
plot.show(g)
```
