```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Week 1/exported_data/googleplaystore_clean.csv")
df
```

```python
df_installs = df.groupby("Category")["Installs"].sum().sort_values(ascending=False).to_frame()
high_score = 33 * -1
df_installs['Score_1'] = range(high_score, len(df_installs) + high_score)
df_installs['Score_1'] = df_installs['Score_1'] * (-1)
df_installs
```

```python
df_price = df.groupby("Category")["Price"].sum().sort_values(ascending=False).to_frame()
high_score = 33 * -1
df_price['Score_2'] = range(high_score, len(df_price) + high_score)
df_price['Score_2'] = df_price['Score_2'] * (-1)
df_price
```

```python
df["Revenue"] = df["Installs"] * df["Price"]
df_revenue = df.groupby("Category")["Revenue"].sum().sort_values(ascending=False).to_frame()
high_score = 33 * -1
df_revenue['Score_3'] = range(high_score, len(df_revenue) + high_score)
df_revenue['Score_3'] = df_revenue['Score_3'] * (-1)
df_revenue
```

```python
df2 = pd.read_csv("../Week 4/Data/googleplaystore_user_reviews_clean_scores.csv")
df2 = df2.drop(labels='Unnamed: 0', axis=1)
df2.head()
```

```python
df1 = pd.merge(df, df2, how='inner', on="App")
df1.head()
```

```python
df_sentiment = df1.groupby(['Category'])['affin_score'].sum().sort_values(ascending=False).reset_index(drop=False)
high_score = 33 * -1
df_sentiment['Score_4'] = range(high_score, len(df_sentiment) + high_score)
df_sentiment['Score_4'] = df_sentiment['Score_4'] * (-1)
df_sentiment
```

```python
df_installs1 = df_installs.drop(labels='Installs', axis=1)
df_price1 = df_price.drop(labels='Price', axis=1)
df_revenue1 = df_revenue.drop(labels='Revenue', axis=1)
df_sentiment1 = df_sentiment.drop(labels='affin_score', axis=1)
```

```python
df_scores = [df_installs1, df_price1, df_revenue1, df_sentiment1]
df_totalScore = pd.merge(df_installs1, df_price1, how="inner", on="Category")
df_totalScore
```

```python
#df_scores = [df_installs1, df_price1, df_revenue1, df_sentiment1]
df_totalScore = pd.merge(df_totalScore, df_revenue1, how="inner", on="Category")
df_totalScore
```

```python
#df_scores = [df_installs1, df_price1, df_revenue1, df_sentiment1]
df_totalScore = pd.merge(df_totalScore, df_sentiment1, how="inner", on="Category")
df_totalScore
```

```python
df_totalScore["Total_Score"] = df_totalScore["Score_1"]+df_totalScore["Score_2"]+df_totalScore["Score_3"]+df_totalScore["Score_4"]
df_totalScore = df_totalScore.sort_values(by="Total_Score", ascending=False).reset_index(drop=True)
df_totalScore
```

```python
sns.set_style("ticks")
plt.figure(figsize=(16,6))
sns.barplot(x="Category", y="Total_Score", data=df_totalScore)
sns.set_context("talk", font_scale=0.7)
plt.title("Category Rankings According To Their Overall Score", fontsize=20)
plt.ylabel("Total Score", fontsize=20)
plt.xlabel("Categories", fontsize=20)
plt.xticks(rotation=90)   
plt.grid(color = 'black', alpha=0.1, linewidth = 2)
plt.show()
plt.clf()
```

```python
sns.set_style("ticks")
plt.figure(figsize=(16,8))
sns.barplot(x="Total_Score", y="Category", data=df_totalScore, orient='horizontal')
sns.set_context("talk", font_scale=0.7)
plt.title("Category Rankings According To Their Overall Score", fontsize=20)
plt.ylabel("Categories", fontsize=20)
plt.xlabel("Total Score", fontsize=20)
#plt.xticks(rotation=0)   
plt.grid(color = 'black', alpha=0.1, linewidth = 2)
plt.show()
plt.clf()
```
